```python
import matplotlib.pyplot as plt

# Data
others = {'Andhra': 25, 'OD': 11, 'TG': 2}
ne_ut = {'AP': 2, 'Assam': 14, 'Goa': 2, 'MAN': 2, 'Meg': 2, 'MIZ': 1, 'NAG': 1, 'Sik': 1, 'TRI': 2, 'ANDNIC': 1, 'CHand': 1, 'DADR': 2, 'laksh': 1}
india = {'Bihar': 25, 'CTTGRH': 5, 'GUJ': 5, 'HAR': 6, 'HP': 2, 'JHK': 7, 'KA': 23, 'KER': 20, 'MP': 10, 'MH': 28, 'OD': 5, 'PB': 10, 'RAJ': 10, 'TN': 40, 'TG': 10, 'UP': 25, 'WB': 20, 'KASH': 3, 'delhi': 5}
nda = {'Bihar': 15, 'CTTGRH': 6, 'GUJ': 21, 'HAR': 4, 'HP': 2, 'JHK': 7, 'KA': 5, 'MP': 19, 'MH': 20, 'del': 2, 'OD': 5, 'PB': 3, 'RAJ': 15, 'TG': 5, 'UP': 55, 'UK': 5, 'WB': 22, 'KASH': 3}

# Function to plot state performance for a given category
def plot_state_performance(data, category_name, top_n=5):
    sorted_states = sorted(data.items(), key=lambda x: x[1], reverse=True)
    top_states = sorted_states[:top_n]
    bottom_states = sorted_states[-top_n:]
    states, seats = zip(*top_states)
    plt.figure(figsize=(12, 6))
    plt.bar(states, seats, color='skyblue')
    plt.xlabel('States')
    plt.ylabel('Number of Seats')
    plt.title(f'Top Performing States in {category_name}')
    plt.xticks(rotation=45)
    plt.show()

    states, seats = zip(*bottom_states)
    plt.figure(figsize=(12, 6))
    plt.bar(states, seats, color='salmon')
    plt.xlabel('States')
    plt.ylabel('Number of Seats')
    plt.title(f'Worst Performing States in {category_name}')
    plt.xticks(rotation=45)
    plt.show()

# Plot top and bottom performing states for India
plot_state_performance(india, "INDIA")

# Plot top and bottom performing states for NDA
plot_state_performance(nda, "NDA")
```

```python
import matplotlib.pyplot as plt

# Function to convert numbers to Indian number system
def indian_number_system(number):
    Indian_number_system = ["", "Hundred", "Thousand", "Lakh", "Crore"]
    i = 0
    while number > 0:
        number, r = divmod(number, 100)
        if r > 0:
            yield f"{r:02d} {Indian_number_system[i]}"
        i += 1

# Data
total_registered_votes = 6300000
total_votes = 46200000
absent_votes = 17000000

# Calculate percentages
absent_percentage = (absent_votes / total_registered_votes) * 100
total_votes_percentage = (total_votes / total_registered_votes) * 100

# Convert total registered votes to Indian number system
total_registered_str = ' '.join(indian_number_system(total_registered_votes))

# Display total registered votes
print(f'Total Registered Votes: {total_registered_str}')

# Plot for the pie chart
plt.figure(figsize=(8, 6))
plt.pie([total_votes, absent_votes], labels=['Total Votes', 'Absent Votes'], autopct='%1.1f%%', startangle=140)
plt.title('Percentage Distribution')

plt.show()
```
