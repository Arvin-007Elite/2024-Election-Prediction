```python
import matplotlib.pyplot as plt

# Data for the pie chart
labels = [
    "DMK (27%)", "Congress (10%)", "VCK (7%)",
    "AIADMK (20%)", "DMDK (2%)",
    "BJP (11%)", "NDA Allies (7%)",
    "NTK (8%)", "Others (4%)", "NOTA (1%)"
]

sizes = [27, 10, 7, 20, 2, 11, 7, 8, 4, 1]  # Percentage values

# Colors for different parties
colors = ["red", "blue", "purple", "green", "lightgreen", "orange", "gold", "yellow", "gray", "pink"]

# Create Pie Chart
plt.figure(figsize=(8, 8))
plt.pie(sizes, labels=labels, autopct='%1.1f%%', colors=colors, startangle=140, wedgeprops={'edgecolor': 'black'})
plt.title("2024 Tamil Nadu Election - Vote Share Distribution")
plt.show()
```

```python
# Correcting the transfer from NDA by splitting into BJP and NDA Allies

# Original vote shares (out of 4.5 Cr votes)
original_votes = {
    "DMK": 27, "Congress": 10, "VCK": 7,
    "AIADMK": 22, "BJP": 11, "NDA Allies": 7,
    "NTK": 8, "Others": 4, "NOTA": 1
}

# Votes transferred to TVK
transfers_to_TVK = {
    "Congress": 5,  # 5% moves to TVK
    "AIADMK": 3,  # 3% moves to TVK
    "NTK": 3,  # 3% moves to TVK
    "BJP": 3,  # 3% moves to TVK
    "NDA Allies": 2  # 2% moves to TVK
}

# New voter turnout increase (from 68% to 75% = +7% new votes to TVK)
new_voter_turnout = 7

# Calculate new vote shares
new_votes = original_votes.copy()
for party, votes in transfers_to_TVK.items():
    new_votes[party] -= votes  # Reduce votes from original party

# Add TVK votes
new_votes["TVK"] = sum(transfers_to_TVK.values()) + new_voter_turnout

# Remove NOTA for better clarity
new_votes.pop("NOTA")

# Pie chart visualization
labels = list(new_votes.keys())
sizes = list(new_votes.values())
colors = ["red", "blue", "purple", "green", "lightgreen", "orange", "gold", "yellow", "gray", "brown"]

plt.figure(figsize=(8, 8))
plt.pie(sizes, labels=labels, autopct='%1.1f%%', colors=colors, startangle=140, wedgeprops={'edgecolor': 'black'})
plt.title("2026 Tamil Nadu Election - Vote Share Distribution (with TVK impact)")
plt.show()
```
